```python
model_checkpoint='facebook/mbart-large-50-many-to-many-mmt'
```

```python
from transformers import MBartForConditionalGeneration,MBart50TokenizerFast
```

```python
model=MBartForConditionalGeneration.from_pretrained(model_checkpoint)
```

```python
tokenizer=MBart50TokenizerFast.from_pretrained(model_checkpoint)
```

```python
tokenizer('My name is Sejal Bhitade')
```

```python
tokenizer(['My name is Sejal Bhitade','I am happy. '])
```

```python
tokenizer("मैं अच्छी तरह से हिंदी नहीं बोलता")
```

```python
text='My name is vedant koli'
```

```python
tokenizer.src_lang='en_XX'
encoded_en=tokenizer(text,return_tensors='pt')
```

```python
encoded_en
```

```python
generated_tokens=model.generate(
    **encoded_en,
    forced_bos_token_id=tokenizer.lang_code_to_id['mr_IN']
)
```

```python
generated_tokens
```

```python
tokenizer.batch_decode(generated_tokens,skip_special_tokens=True)
```

```python
def lang_trans(src='en_XX',out_lang='mr_IN',text=None):
 tokenizer.src_lang=src
 encoded_en=tokenizer(text,return_tensors='pt')

 generated_tokens=model.generate(
    **encoded_en,
    forced_bos_token_id=tokenizer.lang_code_to_id[out_lang]
    )


 return tokenizer.batch_decode(generated_tokens,skip_special_tokens=True)
```

```python
text='''या दिवसात मराठा समाजाला आरक्षण आणि सरसकट कुणबी प्रमाण देण्याची मागणी करण्यात आली आहे.'''
lang_trans(src='mr_IN',out_lang='en_XX',text=text)
```

```python
text='''Who is Manoj Jarange Patil?
Manoj Jarange-Patil, originally from Beed district, settled in Shahgad, Jalna district after getting married. 15 years ago, he actively joined the movement advocating for reservations for the Maratha community in government jobs and education'''
```

```python
lang_trans(src='en_XX',out_lang='hi_IN',text=text)
```
